```python
from sklearn import metrics
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import export_graphviz
import IPython.display
import graphviz
from PIL import Image
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.tree import export_graphviz
import graphviz
from google.colab import drive
drive.mount('/content/drive/')




def kFoldSplit(feat, lab, totSplit, blockInUse, random_state):
    """
    make stratified k fold split of data
    """
    stratifiedKfold = StratifiedKFold(n_splits=totSplit, shuffle=True, random_state=random_state)
    for fold, (trainIndex, valIndex) in enumerate(stratifiedKfold.split(feat, lab)):
        if fold == blockInUse:
            featTrain, featVal = feat[trainIndex], feat[valIndex]
            labelTrain, labelVal = lab[trainIndex], lab[valIndex]
            break
    return featTrain, featVal, labelTrain, labelVal

def calculatefBeta(beta, precision, recall):
    """
    calculate f beta score
    """
    if (precision + recall) == 0:
        return 0
    fBeta = ((1 + beta**2) * (precision * recall)) / ((beta**2 * precision) + recall)
    return fBeta

def oneRoundEvaluation(arborModel, featVal, labelVal, beta=1):
    """
    evaluate performance on single round/single fold
    """
    labelValPredictions = arborModel.predict(featVal)

    confusionMatrix = confusion_matrix(labelVal, labelValPredictions)
    trueNegative, falsePositive, falseNegative, truePositive = confusionMatrix.ravel()

    totalCounts = trueNegative + falsePositive + falseNegative + truePositive
    roundAccuracy = (truePositive + trueNegative) / totalCounts
    if (truePositive + falsePositive) != 0:
        roundPrecision = truePositive / (truePositive + falsePositive)
    else:
        roundPrecision = 0
    if (truePositive + falseNegative) != 0:
        roundRecall = truePositive / (truePositive + falseNegative)
    else:
        roundRecall = 0

    fBeta = calculatefBeta(beta=beta, precision=roundPrecision, recall=roundRecall)

    normRoundFalsePositive = falsePositive / totalCounts
    normRoundFalseNegative = falseNegative / totalCounts
    normRoundTruePositive = truePositive / totalCounts
    normRoundTrueNegative = trueNegative / totalCounts

    summaryDict = {
        "trueNegative": trueNegative,
        "falsePositive": falsePositive,
        "falseNegative": falseNegative,
        "truePositive": truePositive,
        "roundAccuracy": roundAccuracy,
        "roundPrecision": roundPrecision,
        "roundRecall": roundRecall,
        "fBeta": fBeta,
        "labelValPredictions": labelValPredictions,
        "normRoundFalsePositive": normRoundFalsePositive,
        "normRoundFalseNegative": normRoundFalseNegative,
        "normRoundTruePositive": normRoundTruePositive,
        "normRoundTrueNegative": normRoundTrueNegative,
    }

    return summaryDict

def getTBTCI(quantList, confidence=0.95):
    """
    calculate theory based t confidence interval
    """
    xbar1 = np.mean(quantList)
    nn1 = len(quantList)
    degreesFreedom1 = nn1 - 1
    alpha = 1 - confidence
    ssd1 = np.std(quantList, ddof=1)

    tMultiplier = stats.t.ppf(1 - alpha / 2, degreesFreedom1)
    standardError = ssd1 / np.sqrt(nn1)

    lowerBound = xbar1 - tMultiplier * standardError
    upperBound = xbar1 + tMultiplier * standardError
    return [lowerBound, upperBound]

def statAnalysis(kfoldResultNestedData, mlMethod, dataSetInUse, beta=1):
    """
    analyze aggregated results
    """
    allResultsByValue = {
        "trueNegative": [],
        "falsePositive": [],
        "falseNegative": [],
        "truePositive": [],
        "roundAccuracy": [],
        "roundPrecision": [],
        "roundRecall": [],
        "labelValPredictions": [],
        "normRoundFalsePositive": [],
        "normRoundFalseNegative": [],
        "normRoundTruePositive": [],
        "normRoundTrueNegative": [],
        "fBeta": []
    }

    for singleRun in range(len(kfoldResultNestedData)):
        for key in allResultsByValue.keys():
            allResultsByValue[key].append(kfoldResultNestedData[singleRun][key])

    stats_summary = {}
    for key in allResultsByValue.keys():
        if key != "labelValPredictions":
            mean_value = np.mean(allResultsByValue[key])
            ci = getTBTCI(quantList=allResultsByValue[key], confidence=0.95)
            stats_summary[key] = {'mean': mean_value, 'CI': ci}
    return stats_summary

def visualizeTree(arborModel, featureColumns, class_names=['<=50K', '>50K'], max_depth=None, tree_name='tree'):
    """
    graph the tree
    """
    class_names = ['<=50K', '>50K']
    class_names = ['LessThan50K', 'GreaterThan50K']

    dot_data = export_graphviz(
        arborModel,
        out_file=None,
        filled=True,
        rounded=True,
        feature_names=featureColumns,
        class_names=class_names,
        special_characters=False,  # Set to False to avoid issues
        max_depth=max_depth
    )

    graph = graphviz.Source(dot_data)

    graph.format = 'png'
    graph.render(filename=tree_name, cleanup=True)
    IPython.display.display(IPython.display.Image(filename=tree_name + '.png'))


# MAIN CODE

iAmUsingCollab = True
visuals = True
reproSeed = 1
foldCount = 30

df1 = pd.read_csv('/content/drive/My Drive/INTO ML/adultDataset.csv')

df1["income"] = df1["income"].apply(lambda x: x.replace('.', ''))
df1["income"] = df1["income"].apply(lambda x: 1 if x == ">50K" else 0)
df1 = pd.get_dummies(df1, columns=['workclass', 'education', 'marital-status',
                                   'occupation', 'relationship', 'race', 'sex',
                                   'native-country'])

featureColumns = [col for col in df1.columns if col != 'income']
labelColumns = ['income']

featureData = df1[featureColumns].values
labelData = df1[labelColumns].values.ravel()

kfoldResultNestedData_limited = []
kfoldResultNestedData_unlimited = []

for fold in range(foldCount):
    print(f"Processing fold {fold + 1}/{foldCount}")

    featTrain, featVal, labelTrain, labelVal = kFoldSplit(
        feat=featureData,
        lab=labelData,
        totSplit=foldCount,
        blockInUse=fold,
        random_state=reproSeed
    )

    # limited depth tree (max_depth=5)
    arborModel_limited = DecisionTreeClassifier(
        max_depth=5,
        random_state=reproSeed
    ).fit(featTrain, labelTrain)
    result_limited = oneRoundEvaluation(
        arborModel=arborModel_limited,
        featVal=featVal,
        labelVal=labelVal,
        beta=1
    )
    kfoldResultNestedData_limited.append(result_limited)

    # unlimited depth tree (max_depth=None)
    arborModel_unlimited = DecisionTreeClassifier(
        max_depth=None,
        random_state=reproSeed
    ).fit(featTrain, labelTrain)
    result_unlimited = oneRoundEvaluation(
        arborModel=arborModel_unlimited,
        featVal=featVal,
        labelVal=labelVal,
        beta=1
    )
    kfoldResultNestedData_unlimited.append(result_unlimited)


stats_limited = statAnalysis(
    kfoldResultNestedData_limited,
    mlMethod="tree_limited",
    dataSetInUse="adult"
)
stats_unlimited = statAnalysis(
    kfoldResultNestedData_unlimited,
    mlMethod="tree_unlimited",
    dataSetInUse="adult"
)

metrics = ['roundAccuracy', 'roundPrecision', 'roundRecall', 'fBeta',
           "normRoundFalsePositive", "normRoundFalseNegative",
           "normRoundTruePositive", "normRoundTrueNegative"]
differences = {}
for metric in metrics:
    limited_values = [result[metric] for result in kfoldResultNestedData_limited]
    unlimited_values = [result[metric] for result in kfoldResultNestedData_unlimited]
    diff_values = np.array(unlimited_values) - np.array(limited_values)
    mean_diff = np.mean(diff_values)
    ci_diff = getTBTCI(diff_values, confidence=0.95)
    differences[metric] = {'mean_diff': mean_diff, 'CI_diff': ci_diff}

print("\nIndividual Confidence Intervals:")
print("\nLimited Depth Tree:")
for metric in metrics:
    mean = stats_limited[metric]['mean']
    ci = stats_limited[metric]['CI']
    print(f"{metric}: Mean = {mean:.4f}, 95% CI = [{ci[0]:.4f}, {ci[1]:.4f}]")

print("\nUnlimited Depth Tree:")
for metric in metrics:
    mean = stats_unlimited[metric]['mean']
    ci = stats_unlimited[metric]['CI']
    print(f"{metric}: Mean = {mean:.4f}, 95% CI = [{ci[0]:.4f}, {ci[1]:.4f}]")

print("\nConfidence Intervals of the Differences (Unlimited - Limited):")
for metric in metrics:
    mean_diff = differences[metric]['mean_diff']
    ci_diff = differences[metric]['CI_diff']
    print(f"{metric} Difference: Mean = {mean_diff:.4f}, 95% CI = [{ci_diff[0]:.4f}, {ci_diff[1]:.4f}]")

# vis section
print("\nVisualizing the last fold's decision trees:")
if visuals:
    # limited depth tree
    print("Limited Depth Tree (max_depth=5):")
    visualizeTree(
        arborModel_limited,
        featureColumns=featureColumns,
        class_names=['LessThan50K', 'GreaterThan50K'],
        max_depth=None,
        tree_name='limited_tree'
    )

    # unlimited depth tree
    print("Unlimited Depth Tree:")
    visualizeTree(
        arborModel_unlimited,
        featureColumns=featureColumns,
        class_names=['LessThan50K', 'GreaterThan50K'],
        max_depth=5,
        tree_name='unlimited_tree'
    )

```
